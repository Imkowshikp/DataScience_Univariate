```python
def setQuanQual(dataData):
    
    quantitative = [];
    qualitative = [];

    for columnRepresents in dataData.columns:
        
        #print(columnRepresents)
        if(rawData[columnRepresents].dtypes == "O"):
            qualitative.append(columnRepresents);
        else:
            quantitative.append(columnRepresents);
    return quantitative, qualitative;
            
setQuanQual(rawData)
```

```python
Quan, Qual = setQuanQual(rawData);
```

## Calculated Cumulative Frequency

```python
rawData["ssc_p"].cumsum()
```

```python
frequencyTable = pd.DataFrame(index = Quan, columns = ["Frequency", "RelativeFrequency", "CumulativeFrequency"]);

#rowName;
#frequencyTable.drop(["sl_no"], inplace=True)

frequencyTable

for rowName in Quan:
    
    frequencyTable["Frequency"][rowName] = rawData[rowName].value_counts().sort_values(ascending=True)
    frequencyTable["RelativeFrequency"][rowName] = rawData[rowName].value_counts().sort_values(ascending=True)/len(rawData)
    
    
    #relativeValue = rawData[rowName].value_counts().sort_values(ascending=True)/len(rawData)
    
    #frequencyTable["CumulativeFrequency"][rowName] = frequencyTable["RelativeFrequency"].cumsum()
    frequencyTable["CumulativeFrequency"][rowName] = rawData[rowName].cumsum()
```

```python
frequencyTable.drop(["sl_no"], inplace=True)
```
